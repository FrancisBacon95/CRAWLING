```python
from selenium import webdriver
from time import sleep 
import pandas as pd
import datetime
```

```python
driver = webdriver.Chrome('./chromedriver')
```

```python
driver.implicitly_wait(3)
driver.get('https://datalab.naver.com/shoppingInsight/sCategory.naver')
```

```python
topkeyword_perMonth = {}
for y in [2,3]:
    topkeyword_perMonth[y+2016] = {}
    for m in range(1,13):
        topkeyword_perMonth[y+2016][m] = []
```

```python
for y in [2,3]:
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]').click() # 제1분류 선택
    sleep(0.3)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[1]/ul/li[7]').click() # 식품 선택 
    sleep(0.3)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[2]').click() # 제2분류 선택 
    sleep(0.3)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[1]/div/div[2]/ul/li[2]').click() # 수산 선택
    sleep(0.3)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/div').click() # 기간 선택
    sleep(0.3)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/div/ul/li[3]').click() # 월간 선택

    #driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[1]').click() # 시작 년도 선택 

    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/div').click()
    sleep(0.5)
    driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[1]/div/ul/li[3]/a').click()
    sleep(0.5)

    for m in range(1,13):
        month_start = m
        month_end = m+1
        year_start = y
        year_end = y
        
        if datetime.datetime.now().year == year_start+2016 and datetime.datetime.now().month -1 == month_end:
            break
        
        if m == 12:
            month_start = m 
            month_end = 1
            year_start = y
            year_end = y+1
            
            # Ω√¿€≥‚µµ
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[1]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[1]/ul/li['+str(year_start)+']/a').click()
            sleep(0.5)
            # Ω√¿€ ø˘
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[2]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[2]/ul/li['+str(month_start)+']/a').click()
            sleep(0.5)
            # ¡æ∑·≥‚µµ
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[1]/span').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[1]/ul/li[2]/a').click()
            sleep(0.5)
            # ¡æ∑· ø˘
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[2]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[2]/ul/li[1]/a').click()
            sleep(0.5)
            # ¡∂»∏«œ±‚
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a').click()
            
            keywords = driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[2]/div[2]/div/div/div[1]/ul').text
            topkeyword_perMonth[year_start+2016][month_start] = [i[1] for i in enumerate(keywords.split('\n')) if i[0]%2==1][:10]
            sleep(0.2)
            
        else:
            # Ω√¿€≥‚µµ
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[1]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[1]/ul/li['+str(year_start)+']/a').click()
            sleep(0.5)
            # Ω√¿€ ø˘
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[2]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[1]/div[2]/ul/li['+str(month_start)+']/a').click()
            sleep(0.5)
            # ¡æ∑·≥‚µµ
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[1]/span').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[1]/ul/li[1]/a').click()
            sleep(0.5)
            # ¡æ∑· ø˘
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[2]').click()
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/div/div[2]/div[2]/span[3]/div[2]/ul/li[2]/a').click()
            sleep(0.5)
            # ¡∂»∏«œ±‚
            driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[1]/div/a').click()
        
            keywords = driver.find_element_by_xpath('//*[@id="content"]/div[2]/div/div[2]/div[2]/div/div/div[1]/ul').text
            topkeyword_perMonth[year_start+2016][month_start] = [i[1] for i in enumerate(keywords.split('\n')) if i[0]%2==1][:10]
            sleep(0.2)
            
        print(year_start+2016,month_start,year_end+2016,month_end)
    
    sleep(1)
```
