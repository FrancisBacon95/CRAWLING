## 1. konlp의 트위터 형태소 분석기를 이용해서 1차적으로 stopword 제거

```python
from konlpy.tag import *
import time
okt = Okt()
```

```python
# 데이터 입수
text1 = open( '킹크랩 섹션 리뷰.txt',encoding='utf-8').read()
text2 = open( '킹크랩 섹션.txt',encoding='utf-8').read()
# 트위터 nouns 
okt_result11 = okt.nouns(text1)
text11=' '.join(okt_result11)
okt_result21 = okt.nouns(text2)
text22=' '.join(okt_result21)
```

```python
from collections import Counter
import pandas as pd
import numpy as np
```

##  2. stopword list를 이용해서 2차적으로 stopword 제거

```python
stopwords=pd.read_csv("한글불용어.csv",encoding='cp949')
stopword=stopwords["불용어"].tolist()
```

```python
def remove_stopwords(text):
    tokens = text.split(' ')
    stops = stopword
    meaningful_words = [w for w in tokens if not w in stops]
    return ' '.join(meaningful_words)
```

```python
text11_nostop=remove_stopwords(text11)
text22_nostop=remove_stopwords(text22)
```

## 3. TF-IDF 적용

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict

corpus=[text11_nostop,text22_nostop]
```

```python
# ============================================
# -- Get TFIDF
# ============================================
vectorizer = TfidfVectorizer()
sp_matrix = vectorizer.fit_transform(corpus)

word2id = defaultdict(lambda : 0)
for idx, feature in enumerate(vectorizer.get_feature_names()):
    word2id[feature] = idx

i=1
for i, sent in enumerate(corpus):
    print('====== document[%d] ======' % i)
    print( [ (token, sp_matrix[i, word2id[token]]) for token in sent.split() ] )
```
