```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
import sys
import os
if not sys.modules.get('selenium'):
    os.system('pip3 install selenium')
if not sys.modules.get('pandas'):
    os.system('pip3 install pandas')
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
# expected_conditions는 Selenium 2.26.0 이후 부터 사용 가능합니다.
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

```python
year_list2=["2014","2015","2016","2017","2018"]
month_list2=["01","02","03","04","05","06","07","08","09","10","11","12"]
yy_mm_list2=[]
for year in year_list2 :
    for month in month_list2:
        yy_mm=year+month
        yy_mm_list2.append(yy_mm)

print(yy_mm_list2)
```

```python
time_pop=pd.DataFrame(columns=['기준연월','분석영역','00시','01시','02시','03시','04시','05시','06시','07시','08시','09시','10시','11시','12시','13시','14시','15시','16시','17시','18시','19시','20시','21시','22시','23시'])
```

```python
driver=webdriver.Chrome(executable_path="../chromedriver.exe")

id_busan="michinso34"
pd_busan="*wogur95*"
url="https://bigdata.busan.go.kr/sgis/gis/gis.do"
driver.get(url)  # 주소 입력하고 ENTER치는 효과
sleep(5)
elem= driver.find_element_by_id("userid")
elem.send_keys(id_busan)
elem = driver.find_element_by_id("scpd")
elem.send_keys(pd_busan)
elem.send_keys(Keys.RETURN)

driver.get(url)

driver.find_element_by_id("AXselect_AX_sel_sgg_AX_SelectText").click()
```

```python
#서비스인구 탭 클릭
driver.find_element_by_xpath('//*[@id="subMenuContainer"]/li[1]/a').click()
```

```python
driver.switch_to.default_content()
```

```python
for yy_mm in yy_mm_list2 :
    #iframe내로 들어오기
    driver.switch_to_frame(driver.find_element_by_tag_name("iframe"))
    #이렇게 되면 iframe 내에 있는 문서만 읽어져있음.
    # 제대로 읽었는지 확인 : print(driver.page_source)
    print("aa"+yy_mm+"aa")
    driver.find_element_by_id('ipt_year').clear()
    sleep(2)
    driver.find_element_by_id('ipt_year').send_keys(yy_mm)
    
    driver.find_element_by_id("btnAnalysis").click()
    sleep(3)
    #다시 원래 html 문서 전체로 읽어오기
    driver.switch_to.default_content()
    # 원상복구 됐는지 확인 : print(driver.page_source)
    sleep(2)
    #분석 결과가 나와있는 부분도 iframe 태그 내에 있기 때문에
    #다시 그 부분만 읽어오기
    driver.switch_to_frame(driver.find_element_by_id("ifrm_analysisResult"))
    # 제대로 읽었는지 확인 : print(driver.page_source)
    
    sleep(3)
    i_row=1
    #각 행에 대하여
    for row in driver.find_elements_by_tag_name("tr"):   
        if i_row>1 and i_row < 8:
            #print(driver.page_source)
            one_row=[]
            #각 열에 대하여
                
            for col in row.find_elements_by_tag_name("td"):
                one_row.append(col.text)
                
            time_pop.loc[len(time_pop)-1]=one_row
            print(time_pop)
        i_row+=1
    driver.switch_to.default_content()
```

```python
time_pop.to_csv('월평균 시간대별 유동인구.csv',encoding='ms949',index=False)
```
